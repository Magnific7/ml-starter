```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib
from sklearn import tree
```

```python
pd.set_option('display.max_columns',100000)
pd.set_option('display.max_rows',100000)
```

```python
df = pd.read_csv('vgsales.csv')
```

```python
# df
```

```python
df.describe()
```

```python
df = df[df['Year'].notna()]
df.describe()
```

```python
X = df[['NA_Sales','EU_Sales', 'JP_Sales','Other_Sales','Global_Sales','Year']]
y = df['Genre']
```

```python

X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.3)
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
predictions = model.predict(X_test)
# predictions
score = accuracy_score(y_test, predictions)
score
```

# Process and load model

```python

# dump in a file 
# joblib.dump( model,'genre-prediction.joblib')

# Load from stored model
model = joblib.load('genre-prediction.joblib')
predictions = model.predict(X_test)
score = accuracy_score(y_test, predictions)
score
```

# Visualize a decision tree

```python
X = df[['NA_Sales','EU_Sales', 'JP_Sales','Other_Sales','Global_Sales','Year']]
y = df['Genre']
model = DecisionTreeClassifier()
model.fit(X,y)
tree.export_graphviz(model, out_file='genre-recommender.dot',
                 feature_names=['NA_Sales','EU_Sales', 'JP_Sales','Other_Sales','Global_Sales','Year'],
                class_names=sorted(y.unique()),
                 label='all',
                 rounded=True,
                 filled=True)
```

```python
X = df[['NA_Sales','EU_Sales', 'JP_Sales','Other_Sales','Global_Sales','Year']]
y = df['Genre']
model = DecisionTreeClassifier()
model.fit(X,y)
tree_text = tree.export_text(model,
                 feature_names=['NA_Sales','EU_Sales', 'JP_Sales','Other_Sales','Global_Sales','Year'],
                class_names=sorted(y.unique()),
               
                 
                 )
print(tree_text)
```
